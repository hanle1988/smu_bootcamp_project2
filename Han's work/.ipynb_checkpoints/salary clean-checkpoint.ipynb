```python
size = df1.Size.value_counts().index.tolist()
size_counts = df1.Size.value_counts().values.tolist()
colors = ["firebrick", "purple", "pink", "yellowgreen", "lightskyblue", "lightcoral","red","green","yellow"]
explode = (0, 0, 0, 0, 0, 0,0,0,0)
fig = plt.figure(figsize=(10,10))
plt.title("Size of Companies", fontsize = 26, fontweight = "bold", font = "Times New Roman")
plt.axis("equal")
centre_circle = plt.Circle((0,0), 0.70, fc='white')
plt.pie(size_counts, explode=explode, labels=size, colors=colors,
        autopct="%1.1f%%", shadow=False, startangle=40, textprops={'fontsize': 18}, pctdistance=0.85)
fig.gca().add_artist(centre_circle)
plt.show()
```

```python
x = df1["Job Title"].value_counts().values.tolist()
x
```

```python
y = df1["Job Title"].value_counts().index.tolist()
x = df1["Job Title"].value_counts().values.tolist()

df2=pd.DataFrame()
df2["Name"]=y
df2["Number of Jobs"]=x

df2
```

```python
df2.info()
```

```python
df3=df2.loc[df2["Number of Jobs"]>=5]
df3
```

```python

# step 1, get the data
x = df3["Number of Jobs"]
y = df3["Name"]


# step 2, init the plot
fig = plt.figure(figsize=(10,8))
plt.barh(y, x,color="slateblue")

# step 3, customization
plt.title("Job Title", fontsize=20)
plt.xlabel("Number of Jobs", fontsize = 15)
plt.grid(axis="y", alpha=0.25)

# Sets the x limits of the current chart
#plt.xlim(-0.5, 4.5)

# Sets the y limits of the current chart
plt.xlim(0, max(x)+10)
plt.xticks(rotation=0, fontsize = 13)

# step 4, saving/showing
plt.show()
```
