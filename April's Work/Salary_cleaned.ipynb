```python
import pandas as pd
import numpy as np
```

```python
file = "salary_data_cleaned.csv"
df = pd.read_csv(file)
df.head()
```

```python
df.columns
```

```python
df1 = df.rename(columns={"Job Title": "job_title"})
df1.head()
```

```python
df2=df1[["job_title", "Industry","avg_salary","job_state","python_yn", "R_yn", "aws", "excel"]]
df2.head()
```

```python
df2['job_title'].value_counts()
```

```python
df3=df2.groupby('job_title').filter(lambda x : (x['job_title'].count()>=10).any())
df3['job_title'].value_counts()
```

```python
df4=df3[["job_title"]]
df4.info()
```

```python
df5=df4.drop_duplicates(subset=["job_title"], ignore_index=True)
df5
```

```python
df3.to_csv(r'glass.csv', index=False)
```

```python
df5.to_csv(r'job_title.csv', index=False)
```
